```python
import pandas as pd
```

```python
df= pd.read_csv("SOCR-HeightWeight.csv")
```

```python
import plotly.express as px
height=df['Height(Inches)']
weight=df['Weight(Pounds)']
print('the mean of height is', height.mean())
print('the mean of weight is',weight.mean())
print('the mode of height is',height.mode())
print('the mode of weight is',weight.mode())
print('the median of height',height.median())
print('the median of weight',weight.median())
```

```python
import pandas as pd
df1=pd.read_csv('class1.csv')
df2=pd.read_csv('class2.csv')
fig1=px.scatter(df1,x="Student Number",y="Marks",title="class1")
fig2=px.scatter(df2,x="Student Number",y="Marks",title="class2")
marks1=df1["Marks"]
marks2=df2["Marks"]
average1=marks1.mean()
average2=marks2.mean()
print("class1 average",average1)
print("class2 average",average2)
length=marks1.count()
y0=average1
y1=average2
x0=1
x1=length
fig1.update_layout(shapes=[dict(
type="line",y0=y0,y1=y1,x0=x0,x1=x1
)])
fig1.update_yaxes(rangemode="tozero")
fig2.update_yaxes(rangemode="tozero")
fig1.show()
fig2.show()
print("Standard deviation of Marks of class1 from centeral tendancy",marks1.std())
print("Standard deviation of Marks of class from centeral tendancy",marks2.std())
```
